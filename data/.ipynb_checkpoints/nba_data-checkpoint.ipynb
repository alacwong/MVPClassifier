```python
from bs4 import BeautifulSoup
from requests import get
import pandas as pd
```

```python
### Each cell is to play with data so I can get familiar with the format. This way I can parse the data correctly
### when I add it to be database to then be trained

test = 'https://www.basketball-reference.com/leagues/NBA_2016_per_game.html'
response = get(test)
soup = BeautifulSoup(response.text, 'html.parser')

header_str = "Player	Pos	Age	Tm	G	GS	MP	FG	FGA	FG%	3P	3PA	3P%	2P	2PA	2P%\teFG%\tFT	FTA	FT%	ORB	DRB	TRB	" \
          "AST	STL	BLK	TOV	PF	PTS "

# table headers for each stat
headers = [header.strip() for header in header_str.split('\t')]

table = soup.find_all('td')
table_data = [t.text for t in table] # convert table matrix to list

num_rows = int(len(table_data)/len(headers))
row_length = len(headers)

players = []


def fun(x):
    try:
        return float(x)
    except ValueError:
        return x


for i in range(0, len(table_data), row_length):
    # row for player
    row = table_data[i: i+row_length]
    # convert to dict
    player = {headers[i]:fun(row[i]) for i in range(row_length)}
    players.append(player)
    

# # create new headers bc not all labels tell the correct story
# labels = headers[:]
# labels.remove('AGE')
# labels.remove('TM')


pt = pd.DataFrame(players)
test = pt[['Player', 'PTS']]
print(pt)




# find which columns need to be refactored to floats
for column in pt:
    print(column, pt[column].dtype)

print(len(table_data), num_rows, row_length)
print('There are ', len(players), ' in 2015/2016 nba season')
    
    

```

```python
### cell for scraping games won

# response = get('https://www.basketball-reference.com/leagues/NBA_1971_standings.html') 
response = get('https://www.basketball-reference.com/leagues/NBA_2001_standings.html') 

soup = BeautifulSoup(response.text, 'html.parser')

tables = soup.find_all('td')
l = [table for table in tables]


# filtering functions
def east_function(element):
    return [] != element.find_all('th', attrs={'aria-label':"Eastern Conference"})

def west_function(element):
    return [] != element.find_all('th', attrs={'aria-label':"Western Conference"})

# isolate for correct table
east = list(filter(east_function, l))[0] 
west = list(filter(west_function, l))[0]


# setup dictionary for team_link -> winrate for season
wr = list(east.find_all('td', attrs={"data-stat":'win_loss_pct'})) + list(west.find_all('td', attrs={"data-stat":'win_loss_pct'}))
teams = list(east.find_all('a')) + list(west.find_all('a'))
nba = {teams[i]['href']: float(wr[i].text) for i in range(len(wr))}
print(nba)

```

```python
# scrape data for mvp candidates
url = 'https://www.basketball-reference.com/awards/awards_2001.html'
response = get(url)
soup = BeautifulSoup(response.text, 'html.parser')
tables = soup.find_all('table')
l = [table for table in tables]

headers = l[0].find_all('th', attrs={'scope': 'col'})
headers = [h.text for h in headers] # scrape column headers for mvp candidates
headers.remove('Rank')
stats = l[0].find_all('td') # data for mvp candidates

# setup links to teams
team_link = {}
for stat in stats:
    link = stat.find_all('a')
    if link != [] and len(stat.text) == 3:
        team_link[stat.text] = nba[link[0]['href']]
        
stats = [s.text for s in stats] 
mvps = []

for i in range(0, len(stats), len(headers)):
    row = stats[i: i+len(headers)]
    mvp = {headers[j]:row[j] for j in range(len(row))}
    mvps.append(mvp)
df = pd.DataFrame(mvps) # raw table
df['Tm'] = df['Tm'] = df['Tm'].apply(lambda x: team_link[x]) # get winrate
# delete irrelevant fields
del df['G']
del df['Pts Won']
del df['Pts Max']
del df['First']
del df['Age']
print(df)

# we want a subset of the raw data before training the model
# we want to look at important 
```

```python
# cell of cleaing df
print(df)
df2 = df.copy()

# get rid empty cells and
for col in df2:
    count = 0
    for i in range(len(df2[col])):
        if df2[col][i] == '':
            df2.loc[i, col] = '0'
            count+=1
    if count == len(df2[col]):
        del df2[col]

# setup correct types for table
for col in df2:
    df2[col] = pd.to_numeric(df2[col], errors='ignore')

print(df2)


```
